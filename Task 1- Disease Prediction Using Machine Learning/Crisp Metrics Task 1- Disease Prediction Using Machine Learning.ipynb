# **Preparing dataset**

**Split the data to feature and target values**

```python
X = df.iloc[:, 1:].values
Y = df.iloc[:, 0].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.25, random_state = 0)
```

```python
len(X_test)
```

```python
acc = []
model = []
```

# **Training the models**

**SVM CLASSIFIER**

```python
from sklearn.svm import SVC
import sklearn.metrics as metrics
SVM = SVC(gamma='auto')
SVM.fit(X_train,Y_train)
predicted_values = SVM.predict(X_test)
x = metrics.accuracy_score(Y_test,predicted_values)
acc.append(x)
model.append('SVM')
print("Test accuracy : {:2f}%".format(SVM.score(X_test, Y_test) * 100 ))
```

**LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train, Y_train)
predictions = LR.predict(X_test)
x = LR.score(X_test,Y_test)
acc.append(x)
model.append('LR')
print("Test accuracy : {:2f}%".format(LR.score(X_test, Y_test) * 100 ))
```

**RANDOM FOREST CLASSIFIER**

```python
from sklearn.ensemble import RandomForestRegressor
RFC = RandomForestRegressor()
RFC.fit(X_train, Y_train)
accuracy = RFC.predict(X_test)
y = RFC.score(X_test,Y_test)
acc.append(y)
model.append('RFC')
print("Test accuracy : {:2f}%".format(RFC.score(X_test, Y_test) * 100 ))
```

**DECISION TREE CLASSIFIER**

```python
from sklearn.tree import DecisionTreeClassifier
DC = DecisionTreeClassifier()
DC.fit(X_train,Y_train)
tree = DC.predict(X_test)
z = DC.score(X_test,Y_test)
acc.append(z)
model.append('DC')
print("Test accuracy : {:2f}%".format(DC.score(X_test, Y_test) * 100 ))
```

**LINEAR REGRESSION**

```python
from sklearn.linear_model import LinearRegression
LIR = LinearRegression()
LIR.fit(X_train, Y_train)
wild = LIR.predict(X_test)
a = LIR.score(X_test,Y_test)
acc.append(a)
model.append('LIR')
print("Test accuracy : {:2f}%".format(LIR.score(X_test, Y_test) * 100 ))
```

**LINEAR DISCRIMINATE ANALYSIS**

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
LDA = LinearDiscriminantAnalysis()
LDA.fit(X_train, Y_train)
door = LDA.predict(X_test)
b = LDA.score(X_test,Y_test)
acc.append(b)
model.append('LDA')
print("Test accuracy : {:2f}%".format(LDA.score(X_test, Y_test) * 100))
```
